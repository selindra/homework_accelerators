```python
import numpy as np

q = -16e-20
c = 3e8
E = 1e10

# B*rho = E /(q*c)

class particle:
    def __init__(self, q, c, E):
        self.q = q
        self.c = c
        self.E = E 

    def get_rigidity(self):
    
        return E/(q*c)
    
    def __str__(self):
        return str(self.get_rigidity())

#v_rigidity = np.vectorize(get_rigidity)   
v_rigidity = particle(q, c, E)
#a = np.vectorize(v_rigidity)

print('Brho = ', v_rigidity,'[Tm]')
```
